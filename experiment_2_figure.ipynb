```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot([1,2,3], [5.017,4.835,4.637], 's-', label='Potable water',color=(68/255,114/255,196/255))
for a,b in zip([1,2,3],[5.017,4.835,4.637]):
    plt.text(a,b+0.01,'{:.3}'.format(b),ha='center', va='bottom')
plt.plot([1,2,3], [4.234,4.052,3.854], 'o-', label='Non-potable water',color=(237/255,125/255,49/255))
for a,b in zip([1,2,3],[4.234,4.052,3.854]):
    plt.text(a,b+0.01,'{:.3}'.format(b),ha='center', va='bottom')
    
plt.xlabel('Scenarios')
plt.ylabel('Water Consumption (10^5 kilogallon)')
plt.legend(loc='upper right')
plt.xticks([1,2,3],['Base','S3','S4'])

plt.subplot(1,2,2)
plt.plot([1,2,3], [1.887,1.958,2.030], 's-', label='Electric',color=(112/255,173/255,71/255))
for a,b in zip([1,2,3],[1.887,1.958,2.030]):
    plt.text(a,b+0.005,'{:.3}'.format(b),ha='center', va='bottom')

plt.xlabel('Scenarios')
plt.ylabel('Electric Consumption (10^5 MWh)')
plt.legend(loc='lower right')
plt.xticks([1,2,3],['Base','S3','S4'])

plt.savefig('E2.svg', dpi=600, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.bar([0.25,1.85], [5.017,4.986], width=0.3, label='Potable water',color=(68/255,114/255,196/255))
plt.bar([0.55,2.15], [4.234,4.203], width=0.3, label='Non-potable water',color=(237/255,125/255,49/255))
for a,b in zip([0.25,1.85],[5.017,4.986]):
    plt.text(a,b+0.01,'{:.3}'.format(b),ha='center', va='bottom')
for a,b in zip([0.55,2.15],[4.234,4.203]):
    plt.text(a,b+0.01,'{:.3}'.format(b),ha='center', va='bottom')
    
plt.xlabel('Scenarios')
plt.ylabel('Water Consumption (10^5 kilogallon)')
plt.legend(loc='upper center')
plt.xticks([0.4,2.0],['Base','S5'])

plt.subplot(1,2,2)
plt.bar([0.4,1.6], [1.887,1.816], width=0.2, label='Electric',color=(112/255,173/255,71/255))
for a,b in zip([0.4,1.6],[1.887,1.816]):
    plt.text(a,b+0.01,'{:.3}'.format(b),ha='center', va='bottom')

plt.xlabel('Scenarios')
plt.ylabel('Electric Consumption (10^5 MWh)')
plt.legend(loc='upper center')
plt.xticks([0.4,1.6],['Base','S5'])

plt.savefig('E3.svg', dpi=600, bbox_inches='tight')
plt.show()
```

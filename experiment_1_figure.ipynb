```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot([1,2,3], [5.017,4.593,4.938], 's-', label='Potable water',color=(68/255,114/255,196/255))
for a,b in zip([1,2,3],[5.017,4.593,4.938]):
    plt.text(a,b+0.01,'{:.3}'.format(b),ha='center', va='bottom')
plt.plot([1,2,3], [4.234,3.889,4.156], 'o-', label='Non-potable water',color=(237/255,125/255,49/255))
for a,b in zip([1,2,3],[4.234,3.889,4.156]):
    plt.text(a,b+0.01,'{:.3}'.format(b),ha='center', va='bottom')
    
plt.xlabel('Scenarios')
plt.ylabel('Water Consumption (10^5 kilogallon)')
plt.legend(loc='upper center')
plt.xticks([1,2,3],['Base','S1','S2'])

plt.subplot(1,2,2)
plt.plot([1,2,3], [1.887,1.878,1.707], 's-', label='Electric',color=(112/255,173/255,71/255))
for a,b in zip([1,2,3],[1.887,1.878,1.707]):
    plt.text(a,b+0.005,'{:.3}'.format(b),ha='center', va='bottom')

plt.xlabel('Scenarios')
plt.ylabel('Electric Consumption (10^5 MWh)')
plt.legend(loc='upper right')
plt.xticks([1,2,3],['Base','S1','S2'])

plt.savefig('E1.svg', dpi=600, bbox_inches='tight')
plt.show()
```
